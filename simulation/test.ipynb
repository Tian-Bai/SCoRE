```python
import pandas as pd

def get_df(i,j):
    df = pd.read_csv(f'results_conc\\Conc_MDR, case={i}, setting={j}, sigma=0.1, tau=10.0, ncalib=1000, ntest=100, Nrep=100, seedgroup=0, reward=0, oracle=False, fit_Y=False, oracle_mdl=False.csv')
    return df

def get_score(i,j):
    df = pd.read_csv(f'results\\SCoRE_MDR, case={i}, setting={j}, sigma=0.1, tau=10.0, ncalib=1000, ntest=100, Nrep=100, seedgroup=0, reward=0, oracle=False, fit_Y=False, oracle_mdl=False.csv')
    return df

def get_ltt(i,j):
    df = pd.read_csv(f'results_ltt\\LTT_MDR, case={i}, setting={j}, sigma=0.1, tau=10.0, ncalib=1000, ntest=100, Nrep=100, seedgroup=0, reward=0, oracle=False, fit_Y=False, oracle_mdl=False.csv')
    return df

target = 'mdr'
target = 'nsel'
# target = 'reward'

import matplotlib.pyplot as plt

fig, axs = plt.subplots(ncols=3, nrows=2, figsize=(14, 6))
for i in range(3):
    for j in range(2):
        df = get_df(i+1, j+1).groupby(['q', 'tau'], as_index=False).mean()
        df_s = get_score(i+1, j+1).groupby(['q', 'tau'], as_index=False).mean()
        df_l = get_ltt(i+1, j+1).groupby(['q', 'tau'], as_index=False).mean()

        if i == 0 and j == 0:
            axs[j][i].plot(df['q'], df[f'{target}_rad'], label='rad')
            axs[j][i].plot(df['q'], df[f'{target}_hoef'], label='hoef')
            axs[j][i].plot(df['q'], df[f'{target}_naive'], label='naive')
            # axs[j][i].plot(df_l['q'], df_l[f'{target}_ltt'], label='ltt')
            axs[j][i].plot(df_s['q'], df_s[f'{target}'], label='score')
        else:
            axs[j][i].plot(df['q'], df[f'{target}_rad'])
            axs[j][i].plot(df['q'], df[f'{target}_hoef'])
            axs[j][i].plot(df['q'], df[f'{target}_naive'])
            # axs[j][i].plot(df_l['q'], df_l[f'{target}_ltt'])
            axs[j][i].plot(df_s['q'], df_s[f'{target}'])

fig.suptitle(f'{target} plot (MDR case)')
fig.legend()
plt.show()
```
