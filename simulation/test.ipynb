```python
import pandas as pd

def get_df(i,j):
    df = pd.read_csv(f'results_conc\\Conc_MDR, case={i}, setting={j}, sigma=0.1, tau=10.0, ncalib=100, ntest=100, Nrep=100, seedgroup=0, reward=0, oracle=False, fit_Y=False, oracle_mdl=False.csv')
    df = df.groupby(['q', 'tau'], as_index=False).mean()
    return df

def get_score(i,j):
    df = pd.read_csv(f'results\\SCoRE_MDR, case={i}, setting={j}, sigma=0.1, tau=10.0, ncalib=1000, ntest=1000, Nrep=100, seedgroup=0, reward=0, oracle=False, fit_Y=False, oracle_mdl=False.csv')
    df = df.groupby(['q', 'tau'], as_index=False).mean()
    return df

df = get_df(1, 1)
df_s = get_score(1, 1)

import matplotlib.pyplot as plt

plt.plot(df['q'], df['mdr_rad'], label='rad')
plt.plot(df['q'], df['mdr_hoef'], label='hoef')
plt.plot(df['q'], df['mdr_naive'], label='naive')
plt.plot(df_s['q'], df_s['mdr'], label='score')
plt.legend()
plt.show()
```

```python
plt.plot(df['q'], df['reward_rad'], label='rad')
plt.plot(df['q'], df['reward_hoef'], label='hoef')
plt.plot(df['q'], df['reward_naive'], label='naive')

plt.plot(df_s['q'], df_s['reward'], label='score')
plt.legend()
plt.show()
```
